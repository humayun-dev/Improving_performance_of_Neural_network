```python
# Keras tuning/hyperparameters implementation
# how to automate the keras, that we got the best possible approach for the parameters
# the dataset is the diabetes
# Author: Muhammad Humayun Khan

dataset = '/kaggle/input/diabetes/diabetes.csv'
df = pd.read_csv(dataset)
df.head()
```

```python
# extract the X and y
X = df.iloc[:,: -1].values
y = df.iloc[:,-1].values
```

```python
# normalize the values/scale
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X = scaler.fit_transform(X)
```

```python
# train the model
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2, random_state = 1)
```

```python
# import necesseties for the model creation
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense,InputLayer
```

```python
# create a neural network architecture/model
model = Sequential()

model.add(InputLayer(shape = (8,)))
model.add(Dense(32,activation = 'relu'))
model.add(Dense(1,activation = 'sigmoid'))
```

```python
# check summary of the above model
model.summary()
```

```python
# compile the model
model.compile(optimizer = 'Adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
# train the model
history = model.fit(X_train,y_train,batch_size = 32, epochs = 100, validation_data = (X_test,y_test))
```

```python
# keras automation starts from here
# 1. how to select appropriate optimizer
# 2. how to select the neurons or nodes in the layer
# 3. how to select the number of hidden layers in the architecture
# 4. All in one model
```

```python
import keras_tuner as kt
# 1. how to select appropriate optimizer
# create a function, then tuner object and then pass the function to object
def build_model(hp):
    model = Sequential()
    
    model.add(InputLayer(shape = (8,)))
    model.add(Dense(32,activation = 'relu'))
    model.add(Dense(1,activation = 'sigmoid'))
        
    # select the optimizer among all, the string 'optimizer' can be anything of our own choice    
    optimizer = hp.Choice('optimizer',values = ['adam','sgd','rmsprop','adadelta'])
    
    model.compile(optimizer = optimizer,loss = 'binary_crossentropy',metrics = ['accuracy'])
    
    return model
```

```python
# create object of the tuner class and pass to the constructor as the function name, the objective, as what to reduce or increase and 
# here it is validation_accuracy to increase, maximum trials = how many times to repeat

tuner = kt.RandomSearch(build_model,objective = 'val_accuracy',max_trials = 5)
```

```python
# now can use tuner object features
# important feature is search best hyperparameter
tuner.search(X_train,y_train,epochs = 5,validation_data = (X_test,y_test))
```

```python
# get the best tuner among all
# it will return the object and due to [0] can print the best
tuner.get_best_hyperparameters()[0].values

# now we know the best optimizer for our model
```

```python
model = tuner.get_best_models(num_models = 1)[0]
```

```python
model.summary()
```

```python
model.fit(X_train,y_train,validation_data = (X_test,y_test),batch_size = 32, epochs = 100, initial_epoch = 6)
# start from the 6th epoch as first time the epoch was 5 and it was trained
```

```python
# 2. No of neurons needed for a particular layer
# in the above model 32 neurons was in general

def build_model_neurons(hp):
    model = Sequential()
    
    # range[8,128], interval = 8, for first model = 8, 2nd model = 16, 3rd model = 24 so on
    units = hp.Int('units',min_value = 8,max_value = 128,step = 8)  
    model.add(InputLayer(shape = (8,)))
    model.add(Dense(units = units,activation = 'relu'))
    model.add(Dense(1,activation = 'sigmoid'))
    
    model.compile(loss = 'binary_crossentropy', optimizer = 'rmsprop', metrics = ['accuracy'])
    
    return model
```

```python
# create tuner object for the neurons
tuner = kt.RandomSearch(
    build_model_neurons,  # Pass the model-building function
    objective='val_accuracy',  # Define the optimization objective
    max_trials=10,  # Set the number of different hyperparameter combinations to try
    executions_per_trial=1,  # How many times to execute each trial
    directory='my_dir',  # Directory to store the search results
    project_name='hyperparameter_tuning'  # Name of the project
)
```

```python
tuner.search(X_train, y_train, epochs=5, validation_data=(X_test, y_test))
```

```python
# Retrieve the best hyperparameters
tuner.get_best_hyperparameters(num_trials=1)[0].values
```

```python
# Build the best model using the best hyperparameters
model = tuner.get_best_models(num_models = 1)[0]
```

```python
model.summary()
```

```python
model.fit(X_train,y_train,batch_size = 32, epochs = 100, initial_epoch = 6)
```

```python
# how to select number of layers
def build_model_layers(hp):
    model = Sequential()
    
    model.add(InputLayer(shape = (8,)))
    model.add(Dense(72,activation = 'relu'))
    
    for i in range(hp.Int('num_layers',min_value = 1, max_value = 10)):
        model.add(Dense(72,activation = 'relu'))
    model.add(Dense(1,activation = 'sigmoid'))
    
    model.compile(loss = 'binary_crossentropy', optimizer = 'rmsprop', metrics = ['accuracy'])
    
    return model
```

```python
# create tuner object for the neurons
tuner = kt.RandomSearch(
    build_model_neurons,  # Pass the model-building function
    objective='val_accuracy',  # Define the optimization objective
    max_trials=3,  # Set the number of different hyperparameter combinations to try
    executions_per_trial=1,  # How many times to execute each trial
    directory='my_dir_neurons',  # Directory to store the search results
    project_name='hyperparameter_tuning_layers'  # Name of the project
)
```

```python
tuner.search(X_train,y_train,validation_data = (X_test,y_test),epochs = 5)
```

```python
# 4. all in one function
def build_model_all(hp):
    model = Sequential()
    
    # Input Layer
    model.add(InputLayer(shape=(8,)))
    
    # Tune the number of hidden layers and nodes per layer
    for i in range(hp.Int('num_layers', min_value=1, max_value=3)):
        model.add(Dense(units=hp.Int(f'units_{i}', min_value=8, max_value=128, step=8),activation='relu'))
    
    # Output Layer
    model.add(Dense(1, activation='sigmoid'))
    
    # Tune the optimizer
    optimizer = hp.Choice('optimizer', values=['adam', 'sgd', 'rmsprop', 'adadelta'])
    
    model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])
    
    return model
```

```python
# Create a tuner object
tuner = kt.RandomSearch(
    build_model_all,  # Pass the model-building function
    objective='val_accuracy',  # Define the optimization objective
    max_trials=20,  # Set the number of different hyperparameter combinations to try
    executions_per_trial=1,  # How many times to execute each trial
    directory='my_dir_all',  # Directory to store the search results
    project_name='hyperparameter_tuning_all'  # Name of the project
)

```

```python
# Run the hyperparameter tuning
tuner.search(X_train, y_train, epochs=10, validation_data=(X_test, y_test))

```

```python
# Retrieve the best hyperparameters
tuner.get_best_hyperparameters(num_trials=1)[0].values
```
